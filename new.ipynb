```python
def compare_dists(idx):
    hist_neg_y, hist_neg_x = np.histogram(Xj_bg_actual[:,0, idx])
    hist_pos_y, hist_pos_x = np.histogram(Xj_sig_actual[:,0, idx])
    hist_neg_y = hist_neg_y/sum(hist_neg_y)
    hist_pos_y = hist_pos_y/sum(hist_pos_y)
    
    hist_pos_x = (hist_pos_x[:-1])
    hist_neg_x = (hist_neg_x[:-1])
    
    # hist_pos_x.append(hist_pos_x[-1] - hist_pos_x[-2] + hist_pos_x[-1])
    # hist_neg_x.append(hist_neg_x[-1] - hist_neg_x[-2] + hist_neg_x[-1])


    plt.figure(figsize = (4,4))
    plt.ylim((0,1))
    plt.plot(hist_pos_x, hist_pos_y, label='positive')
    plt.plot(hist_neg_x, hist_neg_y, label='negative')
    plt.legend()
    plt.show()
```
