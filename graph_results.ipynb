```python
from glob import glob
import pickle
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
file_list = glob('data/*')
```

```python
all_data_dicts = []

for filename in file_list:
    with open(filename, 'rb') as handle:
        this_dict = pickle.load(handle)
        all_data_dicts.append(this_dict)

```

```python
all_data_dicts
```

```python
from collections import defaultdict
```

```python
results_by_sig_ratio = defaultdict(lambda: [])

for this_dict in all_data_dicts:
    sr = this_dict['sig_ratio']
    results_by_sig_ratio[sr].extend(this_dict['result'])
```

```python
keyss = list(results_by_sig_ratio.keys())
keyss.sort()
keyss
```

```python
results_by_sig_ratio
```

```python
#[-1, 56]
```

```python
def my_graph(points, title):
    
    fig, ax = plt.subplots(figsize = (6,5))

    ax.scatter(*zip(*(points)), color = 'blue', alpha = 0.3)

    ax.set_title(title)
    ax.set_xlabel('columns')
    
    ax.set_ylabel('epochs to converge')
    ax.set_xlim((-20,57))
    ax.set_ylim((-1, 20))
    ax.grid(axis="x", which = 'both')
    
    plt.show()
```

```python
for key in keyss:
    my_graph(results_by_sig_ratio[key], str(key))
```
